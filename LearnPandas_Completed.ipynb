#Working with a large amount of data

```python
for data in pd.read_csv('modified.csv', chunksize = 10):
  print('Chunk DF')
  print(data)
```

```python
new_df = pd.DataFrame(columns = data.columns)

for data in pd.read_csv('modified.csv', chunksize = 5):
  results = data.groupby(['Type 1']).count()
  
  new_df = pd.concat([new_df, results])
```
